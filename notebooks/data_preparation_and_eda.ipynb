```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# 1. Load the enriched data
df = pd.read_csv('../data/enriched_student_data.csv')

# 2. Exploratory Data Analysis (EDA)

# --- Regression Plot: Study Time vs GPA ---
plt.figure(figsize=(10, 6))
sns.regplot(x='StudyTimeWeekly', y='GPA', data=df, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Correlation: Weekly Study Time vs. GPA')
plt.xlabel('Weekly Study Time (Hours)')
plt.ylabel('GPA')
plt.savefig('../plots/unamed.png')
plt.show()

# --- Heatmap: Feature Correlation ---
plt.figure(figsize=(12, 8))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.savefig('../plots/image.png')
plt.show()

# 3. Hypothesis Testing
high_sleep = df[df['SleepHours'] >= 7]['GPA']
low_sleep = df[df['SleepHours'] < 7]['GPA']

t_stat, p_val = stats.ttest_ind(high_sleep, low_sleep)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")

if p_val < 0.05:
    print("Conclusion: Reject the Null Hypothesis. Sleep significantly impacts GPA.")
else:
    print("Conclusion: Fail to reject the Null Hypothesis.")
```
